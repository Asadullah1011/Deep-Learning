Tensorflow Input:
In the last section tensorflow was passed into a session and it returned the result. What if we want to use a non-constant? This is where tf.placeholder() and feed_dict come into place. In this section we will go over the basics of feeding data into Tensorflow



tf.placeholder()
Sadly you can't ust set x to your dataset and put in Tensorflow, beacuse overtime you'll want your Tensorflow model to take in different datasets with different parameters. You need tf.placeholder()!

tf.placeholder() returns a tensor that gets its value from data passed tot he tf.session.run() allowing you to set the input right before the session runs:

Session's feed_dict


```python
x= tf.compat.v1.placeholder(tf.string)

with tf.compat.v1.Session() as sess:
  output = sess.run(x,feed_dict={x: 'test'})
  print(output)
```

```python
x= tf.compat.v1.placeholder(tf.string)
y= tf.compat.v1.placeholder(tf.int32)
z= tf.compat.v1.placeholder(tf.float32)
with tf.compat.v1.Session() as sess:
  output_x= sess.run(x, feed_dict={x: 'Test String', y:123 , z:45.67})
  output_y= sess.run(y, feed_dict={x: 'Test String', y:123 , z:45.67})
  print(output_x)
  print(output_y)
```

```python
x= tf.compat.v1.placeholder("float",None)
y= x ** 2
with tf.compat.v1.Session() as session:
  result= session.run(y, feed_dict={x:[1,2,3]})
print(result)
```

```python
import tensorflow as tf

a = tf.add(2,5)
b = tf.math.multiply(a,3)
#Startup session using default graph
sess= tf.compat.v1.Session()
#Define deictionary that replace the value of a with 15
replace_dict={a:15}
#Run the session, passing in replce_dict as the value to feed_dict
output= sess.run(b, feed_dict=replace_dict)
print(output)
```

# Cross Entropy

```python
import tensorflow as tf

softmax_data= [0.7, 0.2,0.1]
one_hot_data= [1.0, 0.0, 0.0]


softmax= tf.compat.v1.placeholder(tf.float32)
one_hot= tf.compat.v1.placeholder(tf.float32)


#Print cross entropy from session
cross_entropy = -tf.reduce_sum(tf.multiply(one_hot, tf.math.log(softmax)))

with tf.compat.v1.Session() as sess:
  output = sess.run(cross_entropy, feed_dict={one_hot: one_hot_data, softmax: softmax_data})
  print(output)
```
